Import the Libaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('AAPL-Final.csv')
```

```python
df
```

Read the Data

Data Columns

```python
# Changes The Date column as index columns
df.index = pd.to_datetime(df['Date'])
df
```

```python
# drop The original date column
df = df.drop(['Date'], axis='columns')
df
```

Define the Independent Variables

```python
# Create predictor variables
df['Open-Close'] = df.Open - df.Close
df['High-Low'] = df.High - df.Low
```

```python
# Store all predictor variables in a variable X
X = df[['Open-Close', 'High-Low']]
X.head()
```

Define the target variables

```python
# Target variables
#...............syntax: np.where(condition,value_if_true,value_if_false)

y = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)  # We will store +1 for a buy signal and 0 for a no position in y. 
                                                        # If tomorrow’s price is greater than today’s price then we will buy the particular Stock 
y
```

 Split the data into train and test:


```python
split_percentage = 0.8
split = int(split_percentage*len(df))
  
```

```python
# Train data set
X_train = X[:split]
y_train = y[:split]
```

```python
X_train
```

```python
y_train
```

```python
# Test data set
X_test = X[split:]
y_test = y[split:]
```

```python
X_test
```

```python
y_test
```

Support Vector Classifier


```python
# Support vector classifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
cls = SVC().fit(X_train, y_train)
```

```python
#accuracy_score(y_train, y_test)
```

 Strategy implementation


```python
df['Predicted_Signal'] = cls.predict(X)
```

```python
#Calculate daily returns

df['Return'] = df.Close.pct_change()
```

```python
# Calculate strategy returns
df['Strategy_Return'] = df.Return *df.Predicted_Signal.shift(1)
```

```python
# Calculate Cumulutive returns
df['Cum_Ret'] = df['Return'].cumsum()
df
```

```python
# Plot Strategy Cumulative returns 
df['Cum_Strategy'] = df['Strategy_Return'].cumsum()
df
```

strategy vs Cummulative returns

```python
import matplotlib.pyplot as plt
%matplotlib inline
  
plt.plot(df['Cum_Ret'],color='red')
plt.plot(df['Cum_Strategy'],color='blue')
```
